```guile
; base do grafo (compartilhada entre os membros do grupo do projeto TAGAS)

(define graph-schema
    '((directed . 0) (weighted . 1) (direct-loop . 2) (body . 3))
)

(define (getp g p)
    (list-ref g (cdr (assq p graph-schema)))
)

(define (replace-by-i i value l)
    (if (= i 0)
        (cons value (cdr l))
        (cons (car l) (replace-by-i (- i 1) value (cdr l)))
    )
)

(define (setp g p value)
    (replace-by-i (cdr (assq p graph-schema)) value g)
)
```

```guile
(define (get-neighbors V body)
  (let ((row (assoc V body))) 
    (if row
        (cdr row)
        '())))

(define (any-true proc lst)
  (cond ((null? lst) #f)
        ((proc (car lst)) #t)
        (else (any-true proc (cdr lst)))))

(define (can-reach? start target body visited)
  (cond
    ((eq? start target) #t)
    ((member start visited) #f) 
    (else
     (let ((neighbors (get-neighbors start body)))
       (any-true (lambda (v) (can-reach? v target body (cons start visited)))
                 neighbors)))))
```
