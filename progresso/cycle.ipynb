## Sobre a Linguagem

TAGAS é uma linguagem para criação, análise e edição de grafos de forma interativa e dinâmica. Com ela é possível definir pontos e arestas e executar análises estruturais do grafo de forma simples.

## Primitiva: `cycle`

**Sintaxe:** `cycle`  
Não recebe parâmetros. Retorna o **número de ciclos** existentes no grafo corrente.

```python
(define (cycle graph)
  (define count 0)
  
  (define (visita-vizinhos inicio vizinhos visitados)
    (cond
      ((null? vizinhos)
       #f)
      ((equal? (car vizinhos) inicio)
       (set! count (+ count 1))
       (visita-vizinhos inicio (cdr vizinhos) visitados))
      ((and (string>=? (car vizinhos) inicio)
            (not (member (car vizinhos) visitados)))
       (busca inicio (car vizinhos) (cons (car vizinhos) visitados))
       (visita-vizinhos inicio (cdr vizinhos) visitados))
      (else
       (visita-vizinhos inicio (cdr vizinhos) visitados))))

  (define (busca inicio atual visitados)
    (visita-vizinhos inicio (get-neighbors graph atual) visitados))

  (define (percorre-nos nos)
    (cond
      ((null? nos) count)
      (else
       (busca (car nos) (car nos) (list (car nos)))
       (percorre-nos (cdr nos)))))

  (percorre-nos (get-nodes graph)))
```

## Implementação da Primitiva `cycle`

A primitiva percorre o grafo usando **DFS (Busca em Profundidade)** e conta quantos ciclos simples existem.

---

### `count`
Variável que acumula o total de ciclos encontrados durante toda a busca.

---

### `visita-vizinhos`
Percorre a lista de vizinhos do nó atual um a um. Para cada vizinho `w`, três casos são possíveis:
- `w` **é o nó de início** → ciclo encontrado, incrementa `count` e continua com o próximo vizinho
- `w` **é válido** (`>= inicio` e não visitado) → chama `busca` para continuar o caminho a partir de `w`, depois segue para o próximo vizinho
- `w` **é inválido** (já visitado ou menor que o início) → ignora e passa para o próximo

> A condição `string>=? w inicio` garante que cada ciclo seja contado **uma única vez** — sempre a partir do seu menor nó em ordem alfabética.

---

### `busca`
Pega os vizinhos do nó `atual` no grafo e repassa para `visita-vizinhos`. É o passo que avança a DFS um nível mais fundo no caminho.

---

### `percorre-nos`
Itera sobre todos os nós do grafo, usando cada um como possível ponto de início de ciclo. Quando a lista acaba, retorna o valor final de `count`.

---

### `(percorre-nos (get-nodes graph))`
Linha que dispara a execução — pega os nós ordenados e inicia a busca.

```python
; Exemplo 1 
(cycle '(("A1" "A2") ("A2" "A3") ("A3")))           ; 0 - caminho linear

(cycle '(("A1" "A2") ("A2" "A3") ("A3" "A1")))      ; 1 - A1->A2->A3->A1

(cycle '(("A1" "A2") ("A2" "A3" "A1") ("A3" "A1"))) ; 2 - A1->A2->A1
```

```python
; Exemplo 2 - Grafo sem ciclos
(cycle '(("A" "B") ("B" "C") ("C" "D") ("D")))      ; 0 - A->B->C->D
```

```python
; Exemplo 3 - Grafo completo K3
(cycle '(("A" "B" "C") ("B" "A" "C") ("C" "A" "B"))) ; 5
```
